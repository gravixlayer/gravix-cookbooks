# GravixLayer AgentBox - Getting Started

This notebook helps you get started with GravixLayer Sandbox by walking through the simplest possible workflow:
```
create a sandbox → run a “Hello, World” program → view the output.
```
By the end, you’ll have a working sandbox environment and a successful code execution.

## Prerequisites

1. Install the GravixLayer SDK: `pip install gravixlayer`
2. Set your API key as an environment variable: `GRAVIXLAYER_API_KEY`



````python
from dotenv import load_dotenv
load_dotenv()
from gravixlayer import Sandbox
````

````python
sandbox = Sandbox.create(
    template="javascript-base-v1") # By default the sandbox is alive for 5 minutes
result = sandbox.run_code("console.log('Hello World!');",
        language="javascript")
print(result.logs['stdout'][0])
print(f"Created: {sandbox.sandbox_id}")
print(f"Template: {sandbox.template}")
print(f"Timeout: {sandbox.timeout}s")
````
